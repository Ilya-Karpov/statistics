Задание 2. Первичное изучение данных

Проанализировать значения по столбцам, рассчитать базовые статистики, найти выбросы.

```python
short_hd.describe()
```

```python
q1 = short_hd['6_respiratory_rate'].quantile(0.25)
q3 = short_hd['6_respiratory_rate'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = short_hd[short_hd['6_respiratory_rate'].between(lower_bound, upper_bound, inclusive=True)].sort_values('6_respiratory_rate')
remove_outliers
```

```python
# что это за выброс?
pd.concat([short_hd, remove_outliers]).drop_duplicates(keep=False)
```

Выбросы проверил только по respiratory_rate, так как согласно google все остальные параметры вполне нормальны.

Задание 3. Работа с пропусками

Рассчитать количество пропусков для всех выбранных столбцов. Принять и обосновать решение о методе заполнения пропусков по каждому столбцу на основе рассчитанных статистик и возможной взаимосвязи значений в них. Сформировать датафрейм, в котором пропуски будут отсутствовать.

```python
short_hd.isna().sum()
```

```python
filled_hd = pd.read_csv('horse_data.csv', header = None, na_values=['?'])
filled_hd.columns = ['1_surgery?', '2_Age', '3_Hospital Number', '4_rectal_temperature', '5_pulse', '6_respiratory_rate',
                      '7_temperature_of_extremities', '8_peripheral pulse', '9_mucous membranes',  '10_capillary refill time',
                      '11_pain', '12_peristalsis', '13_abdominal distension', '14_nasogastric tube', '15_nasogastric reflux',
                      '16_nasogastric reflux PH', '17_rectal examination', '18_abdomen', '19_packed cell volume',
                      '20_total protein', '21_abdominocentesis appearance', '22_abdomcentesis total protein', '23_outcome', 
                      '24_surgical lesion?','25_type of lesion', '26_type of lesion', '27_type of lesion', '28_cp_data']
filled_hd = filled_hd[['1_surgery?', '2_Age', '4_rectal_temperature', '5_pulse', '6_respiratory_rate', '7_temperature_of_extremities', '11_pain', '23_outcome']]
```

Заменим пропущенное значение в столбцах 1_surgery? и 23_outcome на следующее, так как пропущено всего по одному значению, и разброс невелик, на общую статистику не повлияет.

```python
filled_hd['1_surgery?'].fillna(method='ffill', inplace=True)
filled_hd['23_outcome'].fillna(method='ffill', inplace=True)
```

Заменим пропущенное значение в столбце 4_rectal_temperature на медианное значение, так как взаимосвязей с остальными столбцами не наблюдается.

```python
filled_hd['4_rectal_temperature'].fillna(filled_hd['4_rectal_temperature'].median(), inplace=True)
```

Заменим пропущенное значение в столбце 5_pulse на медианное значение в зависимости от 2_Age.

Исходя из статистики получается, что все последующие параметры так или иначе зависят от 5_pulse, но для разнообразия укажем другие столбцы.

Заменим пропущенное значение в столбце 6_respiratory_rate на медианное значение в зависимости от 2_Age и 1_surgery.

Заменим пропущенное значение в столбце 7_temperature_of_extremities на медианное значение в зависимости от 11_pain.

Заменим пропущенное значение в столбце 11_pain на медианное значение в зависимости от 23_outcome.

```python
filled_hd['5_pulse'].fillna(filled_hd.groupby('2_Age')['5_pulse'].transform('median'), inplace=True)
filled_hd["6_respiratory_rate"].fillna(filled_hd.groupby(["2_Age", "1_surgery?"])["6_respiratory_rate"].transform("median"), inplace = True)
filled_hd['11_pain'].fillna(filled_hd.groupby('23_outcome')['11_pain'].transform('median'), inplace=True)
filled_hd['7_temperature_of_extremities'].fillna(filled_hd.groupby('11_pain')['7_temperature_of_extremities'].transform('median'), inplace=True)
```
